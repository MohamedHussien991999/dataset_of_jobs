```python
Jobs_companies = df['Company'].value_counts()[:5]
print(Jobs_companies)
```

```python
x = df['Title'].value_counts()[:10]
print(x)
```

```python
plt.subplots(figsize=(10,10))
plt.barh(x.index,x,height = 0.5, color='green', edgecolor="black", linewidth=0.1);
plt.title("MOST POPULAR JOBS")
plt.ylabel("JOBS")
plt.xlabel("DEMAND");
```
